```python
# display the regions of CSF, GM WM
# CSF: 1, GM: 2, WM:3
csf_mask = op.get_csf_mask()
gm_mask = op.get_gm_mask()
wm_mask = op.get_wm_mask()
utils.imshow([csf_mask, gm_mask, wm_mask], ["CSF", "GM", "WM"])
```

```python
# Maps
utils.imshow([op.T1_map, op.T2_map, op.PD_map], ["T1", "T2", "PD"])

T1_map_csf = op.get_T1(csf_mask)
T2_map_csf = op.get_T2(csf_mask)
PD_map_csf = op.get_PD(csf_mask)

print("CSF:")
print("T1 CSF = " + str(T1_map_csf) + " ms")
print("T2 CSF = " + str(T2_map_csf) + " ms")
print("PD CSF = " + str(PD_map_csf))

## GM
T1_map_gm = op.get_T1(gm_mask)
T2_map_gm = op.get_T2(gm_mask)
PD_map_gm = op.get_PD(gm_mask)

print("GM:")
print("T1 GM = " + str(T1_map_gm) + " ms")
print("T2 GM = " + str(T2_map_gm) + " ms")
print("PD GM = " + str(PD_map_gm))

## WM
T1_map_wm = op.get_T1(wm_mask)
T2_map_wm = op.get_T2(wm_mask)
PD_map_wm = op.get_PD(wm_mask)

print("WM:")
print("T1 WM = " + str(T1_map_wm) + " ms")
print("T2 WM = " + str(T2_map_wm) + " ms")
print("PD WM = " + str(PD_map_wm))
```

```python
# Pulse sequence
PDw = op.spin_echo(*op.PDw_TRTE)
T1w = op.spin_echo(*op.T1w_TRTE)
T2w = op.spin_echo(*op.T2w_TRTE)

utils.imshow([PDw, T1w, T2w], ["PDw", "T1w", "T2w"])
```
